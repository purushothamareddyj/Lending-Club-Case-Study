```python
# removing the % symbol and convert to float
df_1['int_rate']=df_1['int_rate'].str.replace('%','').astype(float)
df_1['revol_util']=df_1['revol_util'].str.replace('%','').astype(float)
```

```python
# Checking the int_rate column
df_1.int_rate.head()
```

```python
# Remove the Months terms in the terms column and covert to the int64
df_1['term']=df_1['term'].str.replace('months','').astype('int64')
```

```python
# remove < and + symbol and years terms in the emp_lenght column and convert to the int32
df_1['emp_length']=df_1['emp_length'].str.replace('<','').str.replace('+','').str.replace('years','').str.replace('year','').astype('int32')
```
